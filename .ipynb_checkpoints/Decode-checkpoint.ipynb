```python
from scipy.optimize import linear_sum_assignment
from mip import *
import numpy as np
import pandas as pd 
import numpy.random as npr
import itertools
import scipy.special
from collections import Counter
import sys, string, itertools, copy, time
```

Read cell assignment result from the 64 donor experiment

```python
N = 64 # Number of individuals (donors)
P = 12 # Number of pools 
C = 6  # Number of experimental conditions 
S = N*C/P # (Average) number of samples per pool
ct_mtx = np.zeros((N,P), dtype=int) # How many cells in each pool are assigned to each donor
f = "./example/freemux.clust1.samples.counts.mtx"
with open(f,'r') as rf:
    header = next(rf, None)
    for line in rf:
        wd = [int(x) for x in line.strip().split('\t')]
        ct_mtx[wd[1]][wd[0]-1] = wd[2]        
```

Scale cluster sum to 1

```python
code_mat = [] # N x P 
for n in range(N):
    ind_code = ('{0:0%sb}' % int(np.log2(N))).format(n)
    ls_code = [int(x) for x in str(ind_code)]
    # complementary (orthogonal vector)
    ls_code.extend([(x-1)**2 for x in ls_code])
    code_mat.append(ls_code)
    
code_mat = np.array(code_mat, dtype=int)
```

<font size="5">Define cost as the proportioin of non-matched counts</font> 

```python
ct_scale = ct_mtx / np.tile(ct_mtx.sum(axis=1),(P,1)).transpose()
# cost(i,j) = prop. of cluster j assigned to pools not identified by the code if i'th individual
cost = [ [ sum( ct_scale[j]*(1-code_mat[i]) ) for j in range(N) ] for i in range(N)] 
```

```python
ilp = Model(solver_name="CBC")
ilp.verbose = 0
x = [[ilp.add_var(var_type=BINARY) for j in range(N)] for i in range(N) ]
for i in range(N): 
    ilp += xsum(x[i][j] for j in range(N)) == 1
for i in range(N): 
    ilp += xsum(x[j][i] for j in range(N)) == 1
ilp.objective = minimize(xsum( cost[i][j] * x[i][j] for i in range(N) for j in range(N) ) )
status = ilp.optimize(max_seconds=300)
min_cost = ilp.objective_value
ilp.objective_values[0]
```

```python
permut1 = np.array( [ [ x[i][j].xi(0) for j in range(N) ] for i in range(N) ] , dtype=int)
```

```python
permute_sol = []
decode_mat = np.zeros((N,P), dtype=int) # N x P 
pm_vec = [0]*N
for i in range(N):
    j = [k for k,v in enumerate( permut1[i]) if v == 1][0]
    pm_vec[i] = j
    decode_mat[i] = ct_mtx[j]
permute_sol.append([pm_vec])
```

<font size="5"> Original design matrix </font>

```python
code_df = pd.DataFrame(np.transpose(code_mat))
sns.set(font_scale=1.4)
_=plt.figure(figsize=(12,5))
ax = sns.heatmap(code_df, cbar=False)
_=ax.set_yticklabels(ax.get_yticklabels(),rotation=0)
_=plt.xlabel('Individuals')
_=plt.ylabel('Pools')
```

<font size="5"> Decoded matrix </font>

```python
decode_scale = decode_mat / np.tile(decode_mat.sum(axis=1),(P,1)).transpose() * C
code_df = pd.DataFrame(np.transpose(decode_scale))
sns.set(font_scale=1.4)
_=plt.figure(figsize=(15,5))
ax = sns.heatmap(code_df)
_=ax.set_yticklabels(ax.get_yticklabels(),rotation=0)
_=plt.xlabel('Individuals')
_=plt.ylabel('Pools')
```

<font size="5"> Decoded matrix of other solutions (non-optimal) </font> 

```python
ilp = Model(solver_name='CBC')
ilp.verbose = 0
x = [[ilp.add_var(var_type=BINARY) for j in range(N)] for i in range(N) ]
for i in range(N): 
    ilp += xsum(x[i][j] for j in range(N)) == 1
for i in range(N): 
    ilp += xsum(x[j][i] for j in range(N)) == 1
ilp += xsum( cost[i][j] * x[i][j] for i in range(N) for j in range(N) ) - min_cost >= 1e-4
ilp.objective = minimize(xsum( cost[i][j] * x[i][j] for i in range(N) for j in range(N) ) )
status = ilp.optimize(max_seconds=300)
```

```python
permut_list = []
for w in range(len(ilp.objective_values)):
    permut_list.append( np.array( [ [ x[i][j].xi(w) for j in range(N) ] for i in range(N) ] , dtype=int) )
    pm_vec = [0]*N
    decode_mat = np.zeros((N,P), dtype=int) # N x P 
    for i in range(N):
        j = [k for k,v in enumerate( permut_list[w][i]) if v == 1][0]
        pm_vec[i] = j
        decode_mat[i] = ct_mtx[j]
    permute_sol[0].append(pm_vec)
    decode_scale = decode_mat / np.tile(decode_mat.sum(axis=1),(P,1)).transpose() * C
    code_df = pd.DataFrame(np.transpose(decode_scale))
    sns.set(font_scale=1.4)
    _=plt.figure(figsize=(15,5))
    ax = sns.heatmap(code_df)
    _=ax.set_yticklabels(ax.get_yticklabels(),rotation=0)
    _=plt.xlabel('Individuals')
    _=plt.ylabel('Pools')    
```
